Libraries

```python
# Basic Import
import json
import os
import time

from yt_dlp import YoutubeDL
from youtube_transcript_api import YouTubeTranscriptApi
```

## Getting all the videos ids

The following code uses yt-dlp cli to download all the video urls and their respective titles
into a single file called list4.txt.

The code is structured as:

1. --flat-playlist: list all the videos but don't download them
2. --match-title: search for titles that contains "Crítica"
3. --reject-title: search for titles that DOES NOT contains "série"
4. --skip-download: skips the download of videos that matches the reject-title flag
and the ones that DOES NOT match the match-title flag
5. --print-to-file: what are the informations that will be printed to the file. In this case
url and title
6. list4.txt: the name of the file that will have the 
7. last one in the url for the video list from the Youtube Channel

```python
os.system('yt-dlp -s --flat-playlist --match-title "Crítica" --reject-title "série" --skip-download --print-to-file url,title list4.txt "https://www.youtube.com/@phsantos/videos"')
```

```python
with open('list4.txt', 'r') as file:
    lista_video_id = file.read().split('\n')
    lista_video_id = list(map(lambda x: x.split('?v=')[-1], lista_video_id ))
    lista_video_id = lista_video_id[0::2]
```

```python
criticas_list = []

for critica_id in lista_video_id:
    critica_dict = {}

    try: 
        transcricao_sep = YouTubeTranscriptApi.get_transcript(critica_id, languages=['pt'])
        transcricao_junto = ''

        for frase in transcricao_sep:
            transcricao_junto = transcricao_junto + ' ' + frase['text'].replace('[Música]', '')
        
        critica_dict['id'] = critica_id
        critica_dict['texto'] = transcricao_junto
        
        criticas_list.append(critica_dict)

    except Exception as e:
        print(critica_id, e)

    
```

```python
len(criticas_list)
```

```python
with open('criticas.txt', 'w') as file:
    file.write(json.dumps(criticas_list, ensure_ascii=False))
```
